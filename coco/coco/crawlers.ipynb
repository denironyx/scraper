```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_property(property_div):
    property_title_elem = property_div.find('h4', class_='listings-property-title')
    property_title = property_title_elem.text.strip() if property_title_elem else None
    
    location_elem = property_div.find('h4')
    location_text = location_elem.text.strip() if location_elem else None
    
    # Replace property_title in location_text if property_title is not None
    location = location_text.replace(property_title, '') if location_text and property_title else location_text
    
    price_elem = property_div.find('h3', class_='listings-price')
    price = price_elem.text.strip() if price_elem else None
    
    description_elem = property_div.find('p', class_='d-none d-sm-block')
    description = description_elem.text.strip() if description_elem else None
    
    beds_elem = property_div.find('div', class_='fur-areea').find('span')
    beds = beds_elem.text.strip() if beds_elem else None
    
    baths_elem = property_div.find('div', class_='fur-areea').find_all('span')[1]
    baths = baths_elem.text.strip() if baths_elem else None
    
    toilets_elem = property_div.find('div', class_='fur-areea').find_all('span')[2]
    toilets = toilets_elem.text.strip() if toilets_elem else None
    
    agent_phone_elem = property_div.find('div', class_='phone-icon')
    agent_phone = agent_phone_elem.text.strip().replace('Phone', '') if agent_phone_elem else None
    
    return {
        'Property Title': property_title,
        'Location': location,
        'Price': price,
        'Description': description,
        'Beds': beds,
        'Baths': baths,
        'Toilets': toilets,
        'Agent Phone': agent_phone
    }

# Function to scrape data from the website
def scrape_website():
    #response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    property_divs = soup.find_all('div', class_='single-room-text')
    properties_data = [scrape_property(property_div) for property_div in property_divs]
    return properties_data

# URL of the website
#url = "https://www.propertypro.ng/property-for-rent/in/abuja"

# Scrape data from the website
properties_data = scrape_website()

# Create a DataFrame
df = pd.DataFrame(properties_data)

# Display the DataFrame
print(df)
```

```python
import requests
from bs4 import BeautifulSoup, Tag
import pandas as pd
def scrape_property(property_div):
    property_title_elem = property_div.find('h4', class_='listings-property-title')
    property_title = property_title_elem.text.strip() if property_title_elem else None
    
    location_elem = property_div.find('h4')
    location_text = None
    if location_elem:
        # Find the img tag
        img_tag = location_elem.find(alt="icon" ,title="icon")
        if img_tag:
            # Get the text following the img tag
            location_text = img_tag.find_next_sibling(text=True).strip()

    
    price_elem = property_div.find('h3', class_='listings-price')
    price = price_elem.text.strip() if price_elem else None
    
    description_elem = property_div.find('p', class_='d-none d-sm-block')
    description = description_elem.text.strip() if description_elem else None
    
    beds_elem = property_div.find('div', class_='fur-areea').find('span')
    beds = beds_elem.text.strip() if beds_elem else None
    
    baths_elem = property_div.find('div', class_='fur-areea').find_all('span')[1]
    baths = baths_elem.text.strip() if baths_elem else None
    
    toilets_elem = property_div.find('div', class_='fur-areea').find_all('span')[2]
    toilets = toilets_elem.text.strip() if toilets_elem else None
    
    agent_phone_elem = property_div.find('div', class_='phone-icon')
    agent_phone = agent_phone_elem.text.strip().replace('Phone', '') if agent_phone_elem else None
    
    return {
        'Property Title': property_title,
        'Location': location_text,
        'Price': price,
        'Description': description,
        'Beds': beds,
        'Baths': baths,
        'Toilets': toilets,
        'Agent Phone': agent_phone
    }


# Function to scrape data from the website
def scrape_website():
    #response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    property_divs = soup.find_all('div', class_='single-room-text')
    properties_data = [scrape_property(property_div) for property_div in property_divs]
    return properties_data

# URL of the website
#url = "https://www.propertypro.ng/property-for-rent/in/abuja"

# Scrape data from the website
properties_data = scrape_website()

# Create a DataFrame
df = pd.DataFrame(properties_data)

# Display the DataFrame
print(df)
```

```python
# Save DataFrame to a CSV file
df.to_csv('property_data2.csv', index=False)

print("Data saved to property_data.csv")
```
